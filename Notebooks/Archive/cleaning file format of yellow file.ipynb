```python
df1 = pd.read_csv('/content/drive/MyDrive/NYC/2017/yellow_tripdata_2017-01.csv')
df2= pd.read_csv('/content/drive/MyDrive/NYC/2017/yellow_tripdata_2017-02.csv')
df3 =pd.read_csv('/content/drive/MyDrive/NYC/2017/yellow_tripdata_2017-03.csv')
```

```python
for df in [df1,df2,df3]:
  display(df.isnull().sum())
```

```python
for df in [df1,df2,df3]:
  df = df[df.fare_amount>=2.5]
  df = df[df.extra>=0]
  df = df[df.mta_tax>=0]
  df = df[df.tip_amount>=0]
  df = df[df.tolls_amount>=0]
  df = df[df.improvement_surcharge>=0]
  df = df[df.total_amount>=2.5]
  display(df.describe().T)
```

```python
df123=pd.concat([df1,df2,df3],axis=0,ignore_index=True,)
```
